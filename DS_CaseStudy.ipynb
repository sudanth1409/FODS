```python
#How would you handle outliers in the Price column using the interquartile range (IQR) method?
```

```python
q1=df['OrderPrice'].quantile(0.25)
q1
```

```python
q3=df['OrderPrice'].quantile(0.75)
q3
```

```python
IQR=q3-q1
```

```python
lb=q1-1.5*IQR
```

```python
ub=q3+1.5*IQR
```

```python
outlier=((df[['OrderPrice']]<lb) | (df[['OrderPrice']]>ub)).any(axis=1)
```

```python
cleaned=df[~outlier]
```

```python
cleaned
```
